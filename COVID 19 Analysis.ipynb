# Importing Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading The Dataset

```python
df = pd.read_csv('covid_19_india.csv')
df
```

# isnull()

```python
df.isnull().sum()
```

# Fetching the Information

```python
df.info()
```

# Changing the Datatypes

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month Name'] = df['Date'].dt.month_name()
df
```

```python
df['Time'] = df['Time'].str.replace('PM', '')
df['Time'] = df['Time'].str.replace('AM', '')
df
```

```python
df = df.drop(columns = 'Time')
df
```

# Value Counts

```python
sdf = df['State/UnionTerritory'].value_counts().head(10)
sdf = sdf.reset_index()
sdf
```

# Data Visualization

```python
plt.figure(figsize = (18, 5))
plt.bar(sdf['State/UnionTerritory'], sdf['count'])
plt.xlabel('State/Union Territory')
plt.ylabel('Count of Cases in State')
plt.savefig('COVID CASES IN STATES.png', format='png')
```

```python
df
```

```python
mdf = df.groupby('Month Name').agg(ConfirmedDeaths = ('Deaths', 'sum'))
mdf = mdf.reset_index().sort_values(by='ConfirmedDeaths', ascending = False)
mdf
```

# Data Visualization

```python
plt.figure(figsize = (18, 5))
plt.bar(mdf['Month Name'], mdf['ConfirmedDeaths'])
plt.title('Confirmed Deaths in Month')
plt.xlabel('Month Name')
plt.ylabel('Confirmed Deaths in Month')
plt.savefig('COVID DEATHS IN MONTH.png', format='png')
```

```python
df
```

```python
cdf = df.groupby('Month Name').agg(ConfirmedCases = ('Confirmed', 'sum'))
cdf = cdf.reset_index().sort_values(by='ConfirmedCases', ascending = False)
cdf
```

# Data Visualization

```python
plt.figure(figsize = (18, 5))
plt.bar(cdf['Month Name'], cdf['ConfirmedCases'])
plt.title('Confirmed Cases in Month')
plt.xlabel('Month Name')
plt.ylabel('Confirmed Cases in Month')
plt.savefig('COVID CONFIRMED CASES IN MONTH.png', format='png')
```

```python
df
```

```python
tdf = df.groupby('Month Name').agg(CuredCases = ('Cured', 'sum'))
tdf = tdf.reset_index().sort_values(by='CuredCases', ascending = False)
tdf
```

# Data Visualization

```python
plt.figure(figsize = (18, 5))
plt.bar(tdf['Month Name'], tdf['CuredCases'])
plt.title('Cured Cases in Month')
plt.xlabel('Month Name')
plt.ylabel('Cured Cases in Month')
plt.savefig('COVID CURED CASES IN MONTH.png', format='png')
```

```python
df
```

```python
kdf = df.groupby('State/UnionTerritory').agg(TotalDeath = ('Deaths', 'sum'))
kdf = kdf.reset_index().sort_values(by='TotalDeath', ascending=False).head(10)
kdf
```

# Data Visualization

```python
plt.figure(figsize = (18, 5))
plt.bar(kdf['State/UnionTerritory'], kdf['TotalDeath'])
plt.title('Death Cases in State')
plt.xlabel('State/Union Territory')
plt.ylabel('Total Deaths in States')
plt.savefig('COVID DEATH CASES IN STATES.png', format='png')
```

# Kerala Study

```python
udf = df.loc[df['State/UnionTerritory'] == 'Kerala']
udf
```

```python
idf = udf.groupby('Month Name').agg(TotalDeathCases = ('Deaths', 'sum'))
idf = idf.reset_index().sort_values(by='TotalDeathCases', ascending=False)
idf
```

# Data Visualization

```python
plt.figure(figsize = (18, 5))
plt.bar(idf['Month Name'], idf['TotalDeathCases'])
plt.title('Death Cases in Kerala')
plt.xlabel('Month Names')
plt.ylabel('Total Deaths in Kerala')
plt.savefig('COVID DEATH CASES IN KERALA.png', format='png')
```

# Confirmed Cases in Kerala

```python
rdf = udf.groupby('Month Name').agg(TotalConfirmedCases = ('Confirmed', 'sum'))
rdf = rdf.reset_index().sort_values(by='TotalConfirmedCases', ascending=False)
rdf
```

# Data Visualization

```python
plt.figure(figsize = (18, 5))
plt.bar(rdf['Month Name'], rdf['TotalConfirmedCases'])
plt.title('Confirmed Cases in Kerala')
plt.xlabel('Month Names')
plt.ylabel('Total Confirmed Cases in Kerala')
plt.savefig('COVID CONFIRMED CASES IN KERALA.png', format='png')
```
